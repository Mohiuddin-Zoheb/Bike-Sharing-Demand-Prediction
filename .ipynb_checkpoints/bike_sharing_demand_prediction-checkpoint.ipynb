### Handling categorical variables

```python
df.season.value_counts()

#season (1:spring, 2:summer, 3:fall, 4:winter)
```

```python
def season(x):
    
    if x==1:
        return 'spring'
    if x==2:
        return 'summer'
    if x==3:
        return 'fall'
    if x==4:
        return 'winter'
        
```

```python
#appliying season function to convert the numerical values to categorical values

df.season= df.season.apply(season)
```

```python
df['season'].value_counts()
```

```python
df.yr.value_counts()
```

```python
# Converting month column to categorical 

import calendar

df.mnth = df['mnth'].apply(lambda x: calendar.month_abbr[x])
```

```python
df.holiday.value_counts()

# No need to convert into categorical because of binary values
```

```python
df.weekday.value_counts()
```

```python
df.weekday = df['weekday'].map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})
```

```python
df.workingday.value_counts()

# No need to convert into categorical because of binary values
```

```python
df.weathersit.value_counts()

# 1: Clear, Few clouds, Partly cloudy, Partly cloudy
# 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
# 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
# 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
```

```python
df.weathersit = df.weathersit.map({1:'Clear',2:'Mist & Cloudy', 3:'Light Snow',4:'Heavy Rain'})
```

## Creating Dummy variables

```python
df.head()
```

#### We need to create dummy variables for season, month, weekday, weathersit

```python
season_dummy= pd.get_dummies(df['season'], drop_first=True)
```

```python
season_dummy.head()
```

```python
df= pd.concat([df, season_dummy], axis=1)
df.head()
```

```python
# dropping the season column

df= df.drop('season', axis=1)
df.head()
```

```python
month= pd.get_dummies(df['mnth'], drop_first=True)
month.head()
```

```python
df= pd.concat([df, month], axis=1)
df.head()
```

```python
df= df.drop('mnth', axis=1)
df.head()
```

```python
week_day= pd.get_dummies(df['weekday'], drop_first=True)
df= pd.concat([df, week_day], axis=1)
df= df.drop('weekday', axis=1)
df.head()
```

```python
weather= pd.get_dummies(df['weathersit'], drop_first=True)
df= pd.concat([df, weather], axis=1)
df= df.drop('weathersit', axis=1)
df.head()
```

# Splitting the Data into Training and Testing Sets

```python
np.random.seed(0)

df_train, df_test = train_test_split(df, train_size=0.7, test_size=0.3, random_state=100)
```

```python
print(df_train.shape)
print(df_test.shape)
```

# Rescaling the features

```python
scaler = MinMaxScaler()
```

```python
# Performing the rescaling to all columns except dummy variables

num_vars = ['atemp','hum','windspeed','cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
# checking for correlation and plotting a heat map

plt.figure(figsize = (22,22))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

- atemp and cnt are most correlated

```python
# Plotting scatter plot to visualize better

sns.scatterplot(x= df_train.atemp, y= df_train.cnt)
plt.show()
```

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

# Building a linear regression model using statsmodels

```python
import statsmodels.api as sm

# Adding a constant
X_train_lm = sm.add_constant(X_train[['atemp']])

# Create a first fitted model
lr_1 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# checking parameters

lr_1.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.162 + 0.683*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# printing summary

print(lr_1.summary())
```

- p value of consta and atemp is almost 0. the result is statistically significant
- looking at r squared, 41.8% of variance in cnt is explained by atemp


```python
# now lets conside yr variable

X_train_lm = X_train[['atemp', 'yr']]
X_train_lm = sm.add_constant(X_train_lm)
lr_2 = sm.OLS(y_train, X_train_lm).fit()
print(lr_2.summary())
```

### Building model using all the variables

```python
df.columns
```

```python
# using x_train

X_train_lm = sm.add_constant(X_train)
lr_3 = sm.OLS(y_train, X_train_lm).fit()
print(lr_3.summary())
```

- p values of Mar, oct, wed, thursday etc are very high

#### Computing VIF

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# dropping value with high p value(Mar) and low VIF. VIF for all the values will change

X = X_train.drop('Mar', 1)
```

```python
X_train_lm = sm.add_constant(X)

lr_4 = sm.OLS(y_train, X_train_lm).fit()

print(lr_4.summary())
```

```python
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

- VIF of 'workingday' decreased 

```python
# Dropping summer because it is having high pvalue and high VIF

X = X.drop('summer', 1)
X_train_lm = sm.add_constant(X)

lr_5 = sm.OLS(y_train, X_train_lm).fit()

print(lr_5.summary())
```

```python
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Removing Oct since it is having high p value and fairly high VIF

X = X.drop('Oct', 1)
X_train_lm = sm.add_constant(X)

lr_6 = sm.OLS(y_train, X_train_lm).fit()

print(lr_6.summary())
```

```python
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# RFE
### Recursive Feature Elimination

```python
# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm_rfe1 = LinearRegression()
lm_rfe1.fit(X_train, y_train)

# Running RFE

rfe = RFE(lm_rfe1,25)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
print(col)
```

```python
X_train.columns[~rfe.support_]
```

####  Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables

X_train_rfe = X_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm_rfe1 = sm.OLS(y_train, X_train_rfe).fit()
```

```python
lm_rfe1.summary()
```

- we still find some variables with high p value
- lets decrease some more features while performing rfe 
- the current r squared is 0.851

#### decreasing some features and running rfe again

```python
lm_rfe2 = LinearRegression()
lm_rfe2.fit(X_train, y_train)

# Running RFE

rfe = RFE(lm_rfe2,18) # previous was 25
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
print(col)
len(col)
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables

X_train_rfe = X_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm_rfe2 = sm.OLS(y_train, X_train_rfe).fit()
```

```python
lm_rfe2.summary()
```

- after performing with 17 variables, r squared is same
- there are still some columns with high p value namely holiday, Feb and May

#### Calculate the VIFs

```python
vif = pd.DataFrame()
X= X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### Dropping holiday because of high p value 0.103

```python
X_train_new = X_train_rfe.drop(["holiday"], axis = 1)
```

```python
# Building model without holiday

X_train_lm = sm.add_constant(X_train_new)
lm_rfe3= sm.OLS(y_train, X_train_lm).fit()
```

```python
lm_rfe3.summary()
```

- r squared has slightly decreased but Adj. R-squared is the same for both the models
- column May still having high p value

```python
# Dropping 'May' column and rebuilding the model

X_train_new= X_train_new.drop(['May'], axis=1)
```

```python
X_train_lm = sm.add_constant(X_train_new)
```

```python
X_train_lm.head()
```

```python
lm_rfe4 = sm.OLS(y_train,X_train_lm).fit()
```

```python
lm_rfe4.summary()
```

```python
# Dropping 'Feb' column and rebuilding the model

X_train_new= X_train_new.drop(['Feb'], axis=1)
X_train_lm = sm.add_constant(X_train_new)
lm_rfe5 = sm.OLS(y_train,X_train_lm).fit()

lm_rfe5.summary()
```

- lm_rfe5 is the best model with:
    - no columns with high values
    - r squared and Adj. R-squared not getting significantly effected after dropping columns
    - Prob (F-statistic) almost equal to zero

```python
# Calculate the VIFs for the best model

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Residual Analysis

- Residual Analysis is performed to check if the error terms are also normally distributed which is one of the major assumptions of linear regression.
- Let us plot the histogram of the error terms and see what it looks like.

```python
y_train_pred = lm_rfe5.predict(X_train_lm)
```

```python
res= y_train - y_train_pred
fig = plt.figure()
sns.distplot(res)
fig.suptitle('Error Terms', fontsize = 20)            
plt.xlabel('Errors', fontsize = 18) 
```

- The error terms are normally distributed
